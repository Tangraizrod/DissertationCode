```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
from scipy.constants import pi 
from scipy.constants import Boltzmann
```

```python
Luminosity =  3.846e26
```

```python
T = 273.15 #Kelvin
Ds = []
xs =  np.linspace(0, 3.846e26, 100)
for L in xs:
    D = np.sqrt((L)/(16*(T**4)*pi*Boltzmann))
    Ds.append(D)
print(Ds)
#Because we substitute the values in the equation, we put the linspace part inside the x-axis
plt.plot(xs, Ds)
# plt.xscale('log')
# plt.yscale('log')
T = 373.15 #Kelvin
Ds = []
for L in xs:
    D = np.sqrt((L)/(16*(T**4)*pi*Boltzmann))
    Ds.append(D)
print(Ds)
#Because we substitute the values in the equation, we put the linspace part inside the x-axis
plt.plot(xs, Ds)
# plt.xscale('log')
# plt.yscale('log')
```

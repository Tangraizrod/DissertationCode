```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
from scipy.constants import c, h, pi, k
```

```python
T = 5780 #Kelvin
```

```python
lmbda1 = np.logspace(-8., -2, 900)
# print(lmbda1)
```

```python
def Planck(lmbda1):
#     return ((8*pi*h*c)/lmbda1**5)*(1/(np.exp(h*c/(k*T*lmbda1))-1))
    return ((2*h*c**2)/lmbda1**5)*(1/(np.exp(h*c/(k*T*lmbda1))-1))
```

```python
plancks = []
for lmbda in lmbda1:
    planck = Planck(lmbda)
    plancks.append(planck)
```

```python
# print(plancks)
```

```python
plt.plot(lmbda1, plancks)
#plt.scatter(lmbda1, plancks)
plt.xscale("log")
plt.yscale("log")
plt.ylim(1.e-10,1.e+15)
plt.xlabel('Wavelength')
plt.ylabel('Energy radiated per unit volume')
plt.title('Plancks Law plot')
```

```python
IntegratedIntensity = 0
for n in range(1, len(lmbda1)):
    dlmbda = lmbda1[n] - lmbda1[n-1]
    IntegratedIntensity = IntegratedIntensity + plancks[n]*dlmbda
```

```python
print(IntegratedIntensity)
```

```python
Rstar = 696340e3
```

```python
def TotalLuminosity(IntegratedIntensity):
    return 4*(pi**2)*(Rstar**2)*IntegratedIntensity
```

```python
print(TotalLuminosity(IntegratedIntensity))
```
